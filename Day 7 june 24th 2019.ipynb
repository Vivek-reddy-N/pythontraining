```
def printeven(n):
    cnt=0;
    sum=0;
    while(cnt!=n):
        if(cnt%2==0):
          sum=sum+cnt;
        cnt=cnt+1;
    return sum;   
print(printeven(20));
```

```
def l(a):
  sum=0;
  for i in range(len(a)):
    if a[i]%3==0 and a[i]%5==0:
      sum=sum+a[i]
      
  return sum
  
a=[34,45,47,98,97,30]
print(l(a))
```

```
def linearformat(a):
  for i in range(len(a)):
    if i==0 or i==(len(a)-1):
      print(a[i],end=" ")
    else:
      print(a[i-1]*a[i+1],end=" ")
      
a=[1,2,3,4,5]
linearformat(a)
```

```

def linearformat(a):
  for i in range(len(a)):
    if i==0 or i==(len(a)-1):
      print(a[i],end=" ")
    elif a[i-1]%2==0 and a[i+1]%2==0:
      print(a[i],end=" ")
      
a=[1,6,9,4,16,19,22]
linearformat(a)
```

```
def numtolistcon(n):
  list=[]
  while n!=0:
    r=n%10
    list.append(r)
    n=n//10
  list.reverse()
  print(list)
n=int(input("Enter a number"))
print(numtolistcon(n))
```
